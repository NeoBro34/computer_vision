```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Line Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

# Line Style and Color

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title('Styled Line Example')
plt.show()
```

# Multiple Lines in One Graph widh Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')
plt.title('Multiple Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.show()
```

# **Bar Chart with Numpy and Maplotlib**

```python
x = np.array(['Salary','Expense','Charity','Food'])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([5, 7, 8, 7, 2])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y, color='red', s=100)
plt.title('Scatter Plot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

# Histogam with Numpy and Matplotlib

```python
data = np.array([1, 2, 5, 3, 6, 3, 1, 4, 5])

plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

# Subplots(Multiple Graphs Together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.figure(figsize=(8, 4)) # width(wide) | 4 => height(tall)

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Subplot 1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='green')
plt.title('Subplot 2')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data:
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

# plot:
fig, ax = plt.subplots()

ax.plot(x, y, 'o', markersize=2, color='grey')
ax.tripcolor(x, y, z)

ax.set(xlim=(-3, 3), ylim=(-3, 3))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# Draw cuboids in the top left and bottom right corners
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)

# Combine the objects into a single boolean array
voxelarray = cube1 | cube2

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.voxels(voxelarray, edgecolor='k')

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
